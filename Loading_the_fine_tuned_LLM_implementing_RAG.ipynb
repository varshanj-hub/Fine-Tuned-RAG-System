```python
#Load QLoRA Fine-Tuned LLM

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Path to merged fine-tuned model
model_path = "/content/drive/My Drive/fine_tuned_heart_disease_model"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True
)
```

```python
#Load PDF and Split Text

from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

# Path to PDF folder in Drive
pdf_dir = "/content/drive/MyDrive/Heart Diseas RAG- PDFS"
all_docs = []

# Load and split PDFs
for file in os.listdir(pdf_dir):
    if file.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(pdf_dir, file))
        docs = loader.load()
        all_docs.extend(docs)

# Split text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
split_docs = text_splitter.split_documents(all_docs)
```

```python
#Load Embeddings + Create FAISS Vector Store

from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# Use Sentence-Transformers for embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create FAISS vector store
vectorstore = FAISS.from_documents(split_docs, embedding_model)
```

```python
#Build RAG Pipeline with LangChain

from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
from transformers import pipeline

# Create HuggingFace pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512)

# LangChain LLM wrapper
llm = HuggingFacePipeline(pipeline=pipe)

# Create RAG chain
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(search_kwargs={"k": 3}),
    return_source_documents=True
)
```

```python

query = "Why do people get cardiac arrest?"
result = rag_chain(query)

# Extract the final sentence
raw_output = result["result"]

# Optional: If your model returns structured output (like "Answer:", "Context:", etc.), filter it
import re

# Try to remove everything before "Helpful Answer:" or "Answer:"
cleaned = re.split(r"Helpful Answer:|Answer:|\<\/think\>", raw_output, flags=re.IGNORECASE)[-1].strip()

print("📌  Answer:\n")
print(cleaned)
```
